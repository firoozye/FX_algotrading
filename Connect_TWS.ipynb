```python
from ibapi.client import EClient
from ibapi.wrapper import EWrapper
from ibapi.contract import Contract
import threading
import time
import pandas as pd
from datetime import datetime
import pytz
```

```python

class IBapi(EWrapper, EClient):
    def __init__(self):
        EClient.__init__(self, self)
        self.data = []  # Initialize a container to hold our historical data

    def historicalData(self, reqId, bar):
        # Splitting the date and timezone
        date_part, timezone_part = bar.date.rsplit(" ", 1)

        # Parsing the date part
        parsed_date = datetime.strptime(date_part, '%Y%m%d %H:%M:%S')

        # Assigning the timezone
        timezone = pytz.timezone(timezone_part)
        parsed_date = timezone.localize(parsed_date)

        # Append data in the format: Datetime, Open, High, Low, Close
        self.data.append([parsed_date, bar.open, bar.high, bar.low, bar.close])

    def historicalDataEnd(self, reqId, start, end):
        # Indicates the end of historical bars
        print("Historical data download completed")

def run_loop():
    app.run()
```
